```python
from nlp.utils import bars2spacy

"""
input: a ||b|| c ||d|| e
output: [(2,3,'LOC_ACCIDENT'), (4,5,'LOC_ACCIDENT')]
"""

if 'ENT_POS' in df.columns:
    df = df.drop(columns='ENT_POS')
    
df["ENT_POS"] = df['text'].apply(bars2spacy)

df
```

```python
from nlp.utils import bars2y

"""
input: a ||b|| c ||d|| e
output: [0,1,0,1,0]
"""

if 'BARS2Y' in df.columns:
    df = df.drop(columns='BARS2Y')
    
df["BARS2Y"] = df['text'].apply(bars2y)

df
```

```python
from nlp.utils import bars2yLetters

"""
input: aa ||bb|| cb ||db|| eb
output: ---++----++---
"""

if 'BARS2YLETTERS' in df.columns:
    df = df.drop(columns='BARS2YLETTERS')
    
df["BARS2YLETTERS"] = df['text'].apply(bars2yLetters)

df
```
