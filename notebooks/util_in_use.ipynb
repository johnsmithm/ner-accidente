```python
def y2bars(y, text):
    """
    input: 
    - [0,1,0,1,0]
    - a b c d e
    output: a ||b|| c ||d|| e
    """
    text = text.split()
    for index,word in enumerate(zip(y,text)):
        if word[0] == 1:
            text[index] = '||'+ text[index] + '||'
    text = ' '.join(text)
    return text

x = [0,1,0,1,0]
y = 'aa bb cc dd ee'
y2bars(x,y)
```
